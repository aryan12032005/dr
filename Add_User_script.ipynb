```python
# before running this script remember to change the auth in settings.py file for signup view

import requests
import json
def test_signup():
    result=requests.get("http://localhost:8000/get_csrf")
    result=result.json()
    print(result)
    data={
        "email": "",
        "username": "",
        "password":"",
        "first_name": "",
        "last_name": "",
        "phone_number": "",
        "is_faculty": True,
        "is_admin": True,
        "is_allowed": True,
        "dep_code" : ""
    }
    headers= {
          "Content-Type": "application/json",
          "X-CSRFToken": result["csrf_token"],
        }
    result=requests.post("http://localhost:8000/signup/",headers=headers,data=json.dumps(data))
    print(result.json())

def test_login():
    result=requests.get("http://localhost:8000/login")
    result=result.json()
    print(result)

test_signup()
```
